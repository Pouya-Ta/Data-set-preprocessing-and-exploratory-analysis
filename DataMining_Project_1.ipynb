2.1. توجه داشته باشید که فقط مقادیر 1 تا 21 کدهای موجود در جدول ICD را نشان میدهد.از آنجایی که تمام مقادیر موجود در ICD نشاندهنده یک نوع بیماری است،
یک متغیر binary جدید تحت عنوان Disease تعریف کنید که بررسی کند دلیل غیبت  برای هر شخص در بازه ICD قرار دارد یا خیر.
</div>
</font>

```python
def in_icd(val):
    return 1 if 1 <= val <= 21 else 0

df["Disease"] = df["Absence_Reason"].apply(in_icd)
```

<div dir=rtl>
<p align="justify">

</p>
</font>
</div>
</font>

```python
disease_counts = df['Disease'].value_counts()

plt.figure(figsize=(8, 6))
sns.barplot(x=disease_counts.index, y=disease_counts.values, palette="coolwarm")
plt.title('Comparison of Absence Due to Disease vs. Non-Disease Reasons', fontsize=16)
plt.xlabel('Disease (1 = Yes, 0 = No)', fontsize=14)
plt.ylabel('Number of Absences', fontsize=14)
plt.xticks([0, 1], ['Non-Disease', 'Disease'])
plt.show()
```

<b> 2. تحلیل مصرف کنندگان الکل و سیگار </b>
<p align="justify">
<font size=5>
با تجزیه و تحلیل تأثیر مصرف الکل یا دخانیات بر غیبت کارکنان شروع کنیم. با توجه به اینکه سیگار کشیدن و مصرف مکرر الکل تأثیر منفی بر شرایط سلامتی دارند، انتظار داریم که برخی از بیماری‌ها در سیگاری‌ها و مصرف‌کنندگان الکل بیشتر از دیگران باشند.
<br>
1.2. بررسی کنید چند درصد از غیبت ها،از گروه مصرف کننده الکل و چند درصد سیگاری هستند.
</p>
</font>
</div>
</font>

```python
alcohol_absences = df[df['Social_Drinker_Status'] == 'Yes']['Absence_Reason'].count()
total_absences = df['Absence_Reason'].count()
alcohol_absences_percentage = (alcohol_absences / total_absences) * 100

smoker_absences = df[df['Social_Smoker_Status'] == 'Yes']['Absence_Reason'].count()
smoker_absences_percentage = (smoker_absences / total_absences) * 100

print(f"Drinker: {alcohol_absences_percentage:.2f}%")
print(f"Smoker: {smoker_absences_percentage:.2f}%")

```

رابطه بین هزینه‌های حمل و نقل و ساعات غیبت: نمودار جوینت بین هزینه‌های حمل و نقل و ساعات غیبت نیز نشان‌دهنده ارتباط قوی و مشخصی نیست. با این حال، تعدادی از نقاط در محدوده‌های بالاتر هزینه‌های حمل و نقل قرار دارند که ممکن است نشان‌دهنده افزایش غیبت‌ها در بین افرادی با هزینه‌های حمل و نقل بیشتر باشد.

توضیح نمودار جوینت (Joint Plot)
نمودار جوینت یک ابزار بصری است که برای نمایش رابطه بین دو متغیر عددی و توزیع هر یک از آن‌ها استفاده می‌شود. این نمودار ترکیبی از نمودار پراکندگی (scatter plot) و نمودارهای حاشیه‌ای (معمولاً هیستوگرام یا کرنل دانسیتی پلات) است که توزیع هر متغیر را به صورت جداگانه نشان می‌دهد.

نمودار جوینت در موارد زیر کاربرد دارد:

زمانی که می‌خواهید همزمان رابطه بین دو متغیر عددی و توزیع آن‌ها را نمایش دهید.
زمانی که به دنبال بررسی ارتباط بصری بین دو متغیر هستید و همچنین می‌خواهید توزیع هر یک را مشاهده کنید.
زمانی که می‌خواهید نقاط داده‌ای که ممکن است مقادیر بالا یا پایین در هر دو متغیر داشته باشند را شناسایی کنید.
نمودار جوینت می‌تواند برای تحلیل‌های اولیه داده‌ها و شناسایی الگوهای احتمالی در داده‌ها مفید باشد.

همانطور که مشاهده می‌شود، توزیع فواصل از محل اقامت تا محل کار و هزینه‌های حمل و نقل به نظر نزدیک به توزیع‌های نرمال می‌آید، در حالی که زمان غیبت به ساعت‌ها به شدت به سمت راست منحرف شده است. این موضوع باعث می‌شود مقایسه بین متغیرها دشوار باشد. یکی از راه‌حل‌های این مشکل، تبدیل داده‌ها به چیزی نزدیک به توزیع نرمال است. روشی موثر برای انجام این تبدیل، استفاده از تبدیل‌های Box-Cox یا Yeo-Johnson است که هر دو در کتابخانه scipy موجود است.

2.3. با توجه به اینکه ستون زمان غیبت شامل مقادیر صفر است از از تبدیل Yeo-Johnson استفاده کنید.
<br>
سپس هر دو نمودار بالا را مجددا رسم کنید و با تحلیل این این نمودارها درمورد تاثیرگذاری این دو فاکتور بر روی زمان غیبت توضیح دهید.
</p>

```python
from scipy.stats import yeojohnson

df['Absence_Duration_Hours_YJ'], _ = yeojohnson(df['Absence_Duration_Hours'])

sns.jointplot(x='Distance_to_Work', y='Absence_Duration_Hours_YJ', data=df, kind='scatter', color='blue')
plt.title('Relationship Between Distance to Work and Transformed Absence Hours', fontsize=16, pad=70)
plt.xlabel('Distance to Work (km)', fontsize=14)
plt.ylabel('Transformed Absence Hours', fontsize=14)
plt.show()

sns.jointplot(x='Transportation_Expense', y='Absence_Duration_Hours_YJ', data=df, kind='scatter', color='green')
plt.title('Relationship Between Transportation Expense and Transformed Absence Hours', fontsize=16, pad=70)
plt.xlabel('Transportation Expense', fontsize=14)
plt.ylabel('Transformed Absence Hours', fontsize=14)
plt.show()

```
